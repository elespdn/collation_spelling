```python
from collatex import *
collation = Collation()
W1 = open( "data/testElias/A.txt", encoding='utf-8' ).read()
W2 = open( "data/testElias/B.txt", encoding='utf-8' ).read()
W3 = open( "data/testElias/C.txt", encoding='utf-8' ).read()
collation.add_plain_witness( "W1", W1 )
collation.add_plain_witness( "W2", W2 )
collation.add_plain_witness( "W3", W3 )
table = collate(collation, output='html2')
print(table)
```
